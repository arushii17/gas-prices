***GAS PRICES OVER TIME IN USD***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv(r'C:\Users\lenovo\OneDrive\Desktop\PYTHON\gas_prices.csv')

plt.figure(figsize=(8,5))

#plt.plot(gas.Year, gas.USA, 'y.-', label='USA')
#plt.plot(gas['Year'], gas.UK, 'b.-', label='UK')
#plt.plot(gas['Year'], gas['South Korea'], 'r.-', label='South Korea')

for country in gas:
    if country != 'Year' :
        plt.plot(gas.Year, gas[country], marker=('.'), label=country)

plt.xticks(gas.Year[ : :2])

plt.legend(bbox_to_anchor=(1.1, 1.05))     #move legend out of box/graph

plt.xlabel('Year')
plt.ylabel('USD')

plt.title('Gas Prices over time in USD')

plt.savefig('gas prices.png', dpi=150)

plt.show()
```
